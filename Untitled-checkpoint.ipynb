```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
r=requests.get("https://dot.ca.gov/programs/procurement-and-contracts/contracts-out-for-bid")
```

```python
soup=bs(r.content)
print(soup)
```

```python
table_content=soup.find(class_='table')
table_content
```

```python
table=table_content.find_all("td")
table
```

```python
info_list=[]
for tag in table:
    info_list.append(tag.text.strip())
info_list
```

```python
def split(list_a, chunk_size):
    for i in range(0, len(list_a), chunk_size):
        yield list_a[i:i + chunk_size]

info_list=list(split(info_list, 3))
info_list
```
